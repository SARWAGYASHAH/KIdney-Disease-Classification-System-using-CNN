```python
data_dir = "/content/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone"
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import (EarlyStopping, ReduceLROnPlateau,
                                         ModelCheckpoint, LearningRateScheduler)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

IMG_SIZE    = 224
BATCH_SIZE  = 32
NUM_CLASSES = 4
EPOCHS      = 60          # more room to train
LR_INIT     = 1e-3        # start higher, decay down
L2_REG      = 1e-4
```

```python
import os, shutil, random

base_dir = "/content/data"

train_dir = os.path.join(base_dir, "train")
valid_dir = os.path.join(base_dir, "valid")
test_dir  = os.path.join(base_dir, "test")

classes = ["Cyst", "Normal", "Stone", "Tumor"]

# Create folders
for cls in classes:
    os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(valid_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(test_dir, cls), exist_ok=True)

# Split data (IMPORTANT: COPY, NOT MOVE from original)
for cls in classes:
    src = os.path.join(data_dir, cls)
    files = os.listdir(src)
    random.shuffle(files)

    total = len(files)
    train_split = int(0.7 * total)
    valid_split = int(0.85 * total)

    for f in files[:train_split]:
        shutil.copy(os.path.join(src, f), os.path.join(train_dir, cls, f))

    for f in files[train_split:valid_split]:
        shutil.copy(os.path.join(src, f), os.path.join(valid_dir, cls, f))

    for f in files[valid_split:]:
        shutil.copy(os.path.join(src, f), os.path.join(test_dir, cls, f))

print("✅ Clean split done (no leakage)")
```

```python
train_datagen = ImageDataGenerator(
    rescale            = 1./255,
    rotation_range     = 20,          # less rotation for CT scans
    zoom_range         = 0.2,
    shear_range        = 0.15,
    width_shift_range  = 0.1,
    height_shift_range = 0.1,
    horizontal_flip    = True,
    vertical_flip      = False,       # CT scans are orientation-sensitive
    brightness_range   = [0.85, 1.15],
    channel_shift_range = 15.0,       # NEW: subtle color jitter
    fill_mode          = 'reflect'    # CHANGED: avoids black borders
)

# Validation & test: only rescale
valid_datagen = ImageDataGenerator(rescale=1./255)
test_datagen  = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size = (IMG_SIZE, IMG_SIZE),
    batch_size  = BATCH_SIZE,
    class_mode  = 'categorical',
    shuffle     = True
)
valid_data = valid_datagen.flow_from_directory(
    valid_dir,
    target_size = (IMG_SIZE, IMG_SIZE),
    batch_size  = BATCH_SIZE,
    class_mode  = 'categorical',
    shuffle     = False
)
test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size = (IMG_SIZE, IMG_SIZE),
    batch_size  = BATCH_SIZE,
    class_mode  = 'categorical',
    shuffle     = False
)
```

```python
from sklearn.utils.class_weight import compute_class_weight

class_labels = train_data.classes
class_weights_array = compute_class_weight(
    class_weight = 'balanced',
    classes      = np.unique(class_labels),
    y            = class_labels
)
class_weight_dict = dict(enumerate(class_weights_array))
print("Class weights:", class_weight_dict)
```

```python
def conv_block(x, filters, kernel_size=3, strides=1, l2=L2_REG):
    """Conv → BN → ReLU → Conv → BN → ReLU → Pool"""
    x = layers.Conv2D(filters, kernel_size, strides=strides,
                      padding='same',
                      kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)

    x = layers.Conv2D(filters, kernel_size,
                      padding='same',
                      kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)

    x = layers.MaxPooling2D(2, 2)(x)
    return x


def build_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=NUM_CLASSES):
    inputs = layers.Input(shape=input_shape)

    # ── Block 1: 32 filters ─────────────────────────────────
    x = conv_block(inputs, 32)
    x = layers.SpatialDropout2D(0.1)(x)          # spatial dropout early

    # ── Block 2: 64 filters ─────────────────────────────────
    x = conv_block(x, 64)
    x = layers.SpatialDropout2D(0.1)(x)

    # ── Block 3: 128 filters ────────────────────────────────
    x = conv_block(x, 128)
    x = layers.SpatialDropout2D(0.2)(x)

    # ── Block 4: 256 filters ────────────────────────────────  ← NEW
    x = conv_block(x, 256)
    x = layers.SpatialDropout2D(0.2)(x)

    # ── Block 5: 512 filters (light) ────────────────────────  ← NEW
    x = layers.Conv2D(512, 3, padding='same',
                      kernel_regularizer=regularizers.l2(L2_REG))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)

    # ── Global pooling head ─────────────────────────────────
    x = layers.GlobalAveragePooling2D()(x)        # keeps your good choice

    # ── Dense head with L2 + Dropout ────────────────────────
    x = layers.Dense(256, kernel_regularizer=regularizers.l2(L2_REG))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Dense(128, kernel_regularizer=regularizers.l2(L2_REG))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Dropout(0.3)(x)

    # ── Output ──────────────────────────────────────────────
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    return models.Model(inputs, outputs)


model = build_model()
model.summary()
```

```python
loss_fn = tf.keras.losses.CategoricalCrossentropy(
    label_smoothing=0.1    # KEY: prevents overconfident predictions
)

model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=LR_INIT),
    loss      = loss_fn,
    metrics   = ['accuracy']
)
```

```python
def cosine_lr_schedule(epoch, lr):
    """Warm up for 5 epochs, then cosine decay"""
    warmup_epochs = 5
    if epoch < warmup_epochs:
        return LR_INIT * (epoch + 1) / warmup_epochs
    else:
        progress = (epoch - warmup_epochs) / (EPOCHS - warmup_epochs)
        return LR_INIT * 0.5 * (1 + np.cos(np.pi * progress))

lr_scheduler = LearningRateScheduler(cosine_lr_schedule, verbose=1)

early_stop = EarlyStopping(
    monitor              = 'val_accuracy',  # CHANGED: monitor accuracy not loss
    patience             = 12,              # CHANGED: much more patient
    restore_best_weights = True,
    min_delta            = 0.001
)

reduce_lr = ReduceLROnPlateau(
    monitor  = 'val_loss',
    factor   = 0.5,
    patience = 5,           # CHANGED: less aggressive
    min_lr   = 1e-6,
    verbose  = 1
)

checkpoint = ModelCheckpoint(
    'best_kidney_model.keras',
    monitor              = 'val_accuracy',
    save_best_only       = True,
    verbose              = 1
)

# NOTE: Use lr_scheduler OR reduce_lr, not both simultaneously
# lr_scheduler gives more control; use it as primary
callbacks = [early_stop, lr_scheduler, checkpoint]
```

```python
history = model.fit(
    train_data,
    validation_data = valid_data,
    epochs          = EPOCHS,
    callbacks       = callbacks,
    class_weight    = class_weight_dict    # handles imbalance
)
```
